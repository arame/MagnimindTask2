```python
import pandas as pd
import numpy as np
```

```python
def set_dataframe_data_types(df_recommender):
    df_customer_ids = df_recommender["Customer_ID"].astype("string").to_frame()
    df_products = df_recommender.drop(["Customer_ID"], axis=1)
    df_products = df_products.astype('Int64')
    df_new_recommender = df_products.assign(Customer_ID = df_customer_ids)
    return df_customer_ids, df_products, df_new_recommender
```

```python
file_folder = "Files"
file_name = "recom_pivot.csv"
path = f"{file_folder}//{file_name}"
df_recommender = pd.read_csv(path, header=0)
df_customer_ids, df_products, df_recommender = set_dataframe_data_types(df_recommender)
df_products.head()
```

```python
#selected_product = "200019.5"
#selected_product = "5001.5"
#selected_product = "5002.0"
#selected_product = "5000.5"
selected_product = "5001.0"
df_corr = df_products.corr()
df_single_product = df_corr[selected_product].to_frame()
df_sort = df_single_product.fillna(0).sort_values(by=[selected_product], ascending=False)
print(f"If you have bought {selected_product} then you may want to consider the following products:")
df_sort.head()


```
