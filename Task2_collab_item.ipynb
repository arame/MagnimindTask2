This code performs a collaborative based item recommendation system.
This approach is a memory based solution

```python
import pandas as pd
import numpy as np
from product_data import ProductData
```

```python
file_name = "recom_pivot.csv"
prod = ProductData(file_name)
prod.set_pivot_dataframe_data_types()
prod.df_orig_recommender.head()
prod.df_products.head()
```

```python
def output_results(selected_product, df_corr):
    df_single_product = df_corr[selected_product].to_frame()
    df_sort = df_single_product.fillna(0).sort_values(by=[selected_product], ascending=False)
    print(f"If you have bought {selected_product} then you may want to consider the following products:")
    print(df_sort.head())
    print("--------------------------")


```

```python
selected_products = ["200019.5", "5001.0", "5001.5", "5002.0", "5000.5"]
df_corr = prod.df_products.corr()
for selected_product in selected_products:
    output_results(selected_product, df_corr)
```

If a person buys product 5001.0, then the recommender systems suggests that that person may want to also buy the products 5000.5, 10023.5, 5002.0, 10006.0. These products have a low selling correlation with product 5001.0 of less than 0.2 and may not be very good choices. The same applies (ie weak correlation) for the other products tested.
